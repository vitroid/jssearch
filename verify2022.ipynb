```python
# プログラム編成後の情報と、こちらでもっている情報を比較する。
# こちらに反映させるべき。
# 内容が変更されると、PDFも再生成する必要がある。あ、できるか。

import json
import sys
from difflib import SequenceMatcher
from subprocess import STDOUT, TimeoutExpired, check_output
import glob

import pandas as pd


def similar(a, b):
    return SequenceMatcher(None, a, b).ratio()


official = pd.read_excel("プログラム編成用_72_7f_松本先生送付用 発表申込 ALL.xlsx")

data = dict()
for j in glob.glob("master/*.json"):
    with open(j) as f:
        d = json.load(f)
        # d.read_json(j, orient="index")
        data[d["id"]] = d

data = pd.DataFrame.from_dict(data, orient="index")    
# 何が対応しているべきか。まず1つだけ抽出してみる。
# print(official[official["id"] == 72500])
# print(data[data["id"] == 72500])
# print(data)

data
```

```python
# 表が大きすぎて扱いにこまる。今調べたいのは、とりあえず題名と講演者だけなので、
# それらだけを抽出する。

minidata = data[['id', 'title', 'titlee']]
minidata = minidata.astype({'id': 'int32'})
minidata = minidata.set_index('id')

minidata
```

```python
miniofficial=official[["id", "講演題目\n7", "Title of the talk\n８"]]
miniofficial = miniofficial.dropna(subset=['id'])
miniofficial = miniofficial.astype({'id': 'int32'})
miniofficial = miniofficial.set_index('id')
miniofficial
```

```python
import numpy as np

merged = minidata.join(miniofficial)
# 英語題名にひらがなを含むものをさがす。
ser1 = merged["titlee"].str.contains(r"[あ-ん]")
merged[ser1]
```

```python
# master.jsonの日英題名を入れ替える。

with open("master.json") as f:
    master = json.load(f)

for key, rec in master.items():
    key = int(key)
    if ser1.loc[key]:
        rec["title"], rec["titlee"] = rec["titlee"], rec["title"]

with open("master-v1.json", "w") as f:
    json.dump(master, f, indent=4, sort_keys=True, ensure_ascii=False)



                
```
